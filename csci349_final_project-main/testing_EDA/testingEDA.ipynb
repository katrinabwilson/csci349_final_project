````python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# List of image paths
image_paths = ['data/monet_jpg/000c1e3bff.jpg', 'data/monet_jpg/011835cfbf.jpg', 'data/monet_jpg/0260d15306.jpg']


# Initialize lists to store statistics
mean_values = []
median_values = []

for path in image_paths:
    # Load the image
    image = cv2.imread(path)
    
    # Convert BGR to RGB
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    # Extract individual color channels
    red_channel = image_rgb[:, :, 0]
    green_channel = image_rgb[:, :, 1]
    blue_channel = image_rgb[:, :, 2]
    
    # Calculate mean and median for each color channel
    red_mean = np.mean(red_channel)
    green_mean = np.mean(green_channel)
    blue_mean = np.mean(blue_channel)
    
    red_median = np.median(red_channel)
    green_median = np.median(green_channel)
    blue_median = np.median(blue_channel)
    
    # Store statistics
    mean_values.append((red_mean, green_mean, blue_mean))
    median_values.append((red_median, green_median, blue_median))

# Convert lists to numpy arrays for easier manipulation
mean_values = np.array(mean_values)
median_values = np.array(median_values)

# Create DataFrames for seaborn
import pandas as pd

# For means
data_mean = pd.DataFrame(mean_values, columns=['Red', 'Green', 'Blue'])
# data_mean['Statistic'] = 'Mean'
# data_mean['Image'] = ['Image ' + str(i+1) for i in range(len(image_paths))]
data_mean

# Plot the means
sns.scatterplot(data=data_mean)
plt.show()
# For medians
# data_median = pd.DataFrame(median_values, columns=['Red', 'Green', 'Blue'])
# data_median['Statistic'] = 'Median'
# data_median['Image'] = ['Image ' + str(i+1) for i in range(len(image_paths))]

# # Combine dataframes
# data = pd.concat([data_mean, data_median], ignore_index=True)
# # Plotting means
# plt.figure(figsize=(10, 6))
# sns.scatterplot(data=data_mean, x='Red', y='Green', hue='Blue', palette=data_mean[['Red', 'Green', 'Blue']], legend='brief')
# plt.xlabel('Red Intensity (Mean)')
# plt.ylabel('Green Intensity (Mean)')
# plt.title('Mean Pixel Intensity Across Color Channels')
# plt.grid(axis='both')
# plt.legend(title='Blue Intensity')
# plt.show()

# # # Plotting medians
# # plt.figure(figsize=(10, 6))
# # sns.scatterplot(data=data_median, x='Red', y='Green', hue='Blue', palette=data_median[['Red', 'Green', 'Blue']], legend='brief')
# # plt.xlabel('Red Intensity (Median)')
# # plt.ylabel('Green Intensity (Median)')
# # plt.title('Median Pixel Intensity Across Color Channels')
# # plt.grid(axis='both')
# # plt.legend(title='Blue Intensity')
# # plt.show()

# # Print results
# print("Overall Mean:", np.mean(mean_values, axis=0))
# print("Overall Median:", np.median(median_values, axis=0))
````
